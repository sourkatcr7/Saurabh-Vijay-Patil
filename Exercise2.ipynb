```python
import seaborn as sns
import pandas as pd


iris = sns.load_dataset('iris')


print(iris.head())
```

The Iris dataset consists of 150 rows with 5 columns: sepal_length, sepal_width, petal_length, petal_width, and species.
We will use seaborn.load_dataset('iris') to load the data into a pandas DataFrame.


```python
#2. Simple Random Sampling:
import numpy as np


np.random.seed(42)


sampled_data = iris.sample(n=30)


print(sampled_data.head())
```

We use iris.sample(n=30) to randomly select 30 observations from the Iris dataset.
We set the random seed (np.random.seed(42)) to ensure that the sampling is reproducible.

```python
#3. Sample Mean Distribution Analysis (Using CLT):
import matplotlib.pyplot as plt


num_samples = 100
sample_means = []


for _ in range(num_samples):
    sampled_data = iris.sample(n=30)
    sample_means.append(sampled_data['sepal_length'].mean())


plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of 100 Sample Means of Sepal Length')
plt.xlabel('Sample Mean of Sepal Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

We repeat the random sampling process 100 times, each time taking a sample of 30 observations and calculating the mean of the sepal_length column.
We store the sample means in the sample_means list.
After collecting all the sample means, we plot a histogram to visualize the distribution of the sample means.
According to the Central Limit Theorem, the distribution of the sample means should approach a normal distribution as the number of samples increases, even if the original population distribution is not normal.

```python
#4. Systematic Sampling:

def systematic_sampling(df, sample_fraction=0.2):

    k = int(len(df) * sample_fraction)


    sampled_data = df.iloc[::k]
    return sampled_data


systematic_sample = systematic_sampling(iris, sample_fraction=0.2)


plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)
plt.scatter(iris['sepal_length'], iris['sepal_width'], color='blue', label='Original Data')
plt.title('Original Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


plt.subplot(1, 2, 2)
plt.scatter(systematic_sample['sepal_length'], systematic_sample['sepal_width'], color='red', label='Systematic Sample')
plt.title('Systematic Sampling (20%)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.tight_layout()
plt.show()
```
