```python
import pandas as pd


file_name = 'student_lifestyle_dataset.csv'


data = pd.read_csv(file_name)


print(data.head())


print(data.info())
```

```python
# Step 3: Perform descriptive statistical analysis

numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns


for col in numerical_columns:
    print(f"Statistics for {col}:")
    print(f"  Mean: {data[col].mean():.2f}")
    print(f"  Median: {data[col].median():.2f}")
    print(f"  Variance: {data[col].var():.2f}")
    print(f"  Standard Deviation: {data[col].std():.2f}")
    print(f"  Range: {data[col].max() - data[col].min():.2f}")
    print("-" * 50)
```

```python

print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns


sns.set(style="whitegrid")


for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[col], kde=True, bins=20, color="skyblue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Step 6: Boxplot for GPA vs Stress Level
plt.figure(figsize=(8, 6))
sns.boxplot(x="Stress_Level", y="GPA", data=data, palette="coolwarm")
plt.title("GPA by Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("GPA")
plt.show()
```

```python
# Step 7: Correlation heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = data[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="viridis", fmt=".2f", square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 8: Scatterplot of GPA vs Study Hours
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Study_Hours_Per_Day", y="GPA", data=data, hue="Stress_Level", palette="coolwarm")
plt.title("GPA vs Study Hours Per Day")
plt.xlabel("Study Hours Per Day")
plt.ylabel("GPA")
plt.show()
```

```python
#Here are some key insights based on the results of descriptive analysis and visualizations from the datasets
#1. Study Hours and GPA
#Insight:

#If Study_Hours_Per_Day has a positive correlation with GPA (from the heatmap or scatterplot), it indicates that students who study more tend to have higher academic performance.
sns.scatterplot(x="Study_Hours_Per_Day", y="GPA", data=data, hue="Stress_Level", palette="coolwarm")
```
